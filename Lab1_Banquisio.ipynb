# SORT() & SORTED () <hr style="border: none; border-top: 1px dashed #ff56bb;width: 75%; margin: 10px 0;"></hr>

```python
# SORT () - mutates the list
numbers = [20, 5, 15, 10]
numbers.sort()
print(numbers)
```

```python
# SORTED () - returns new list but immutable
numbers = [20, 5, 15, 10]
sorted(numbers)
```

# TUPLE <hr style="border: none; border-top: 1px dashed #ff56bb;width: 75%; margin: 10px 0;"></hr>

```python
# JUST LIKE LIST BUT DENOTED BY PARENTHESIS () 
# TUPLES ARE IMMUTABLE

mixed_list = ('white','fruits', 10, 'Ken','ILYS1892')
print(mixed_list)
```

```python
mixed_list = ('white', 'fruits', 10, 'Ken', 'ILYS1892')

# 1. INDEXING -------------------------
print("Indexing:")
print("Index [0]:", mixed_list[0])
print("Index [2]:", mixed_list[2])

# 2. MODIFYING ELEMENTS -------------------------
temp_list = list(mixed_list)
temp_list[0] = 'vegetables'
mixed_list = tuple(temp_list)
print("\nAfter modifying:")
print(mixed_list)  # tuple output -> ( )

# 3. ADDING ELEMENTS -------------------------
temp_list = list(mixed_list)
temp_list += ('blue',)  # add element (as tuple)
mixed_list = tuple(temp_list)
print("\nAfter adding an element:")
print(mixed_list)

# 4. REMOVING ELEMENTS -------------------------
temp_list = list(mixed_list)
temp_list.remove('Ken')
mixed_list = tuple(temp_list)
print("\nAfter removing an element:")
print(mixed_list)

# 5. SORTED -------------------------
# sorted() returns a list, so convert back to tuple
mixed_list = tuple(sorted(mixed_list, key=str))
print("\nAfter sorting:")
print(mixed_list)

# 6. SLICING -------------------------
print("\nSlicing tuple:")
print("mixed_list[1:4]:", mixed_list[1:4])
print("mixed_list[:3]:", mixed_list[1:3])
print("mixed_list[2:]:", mixed_list[2:3])
```

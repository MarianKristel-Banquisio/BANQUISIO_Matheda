```python
import numpy as np
from scipy import stats

# DATA BASED ON THE ASSIGNMENT ----------------------------------------------------------
Jaguar = [10.6, 9.1, 9.5, 9.8, 10.5, 10.4, 9.5, 11, 10.4, 9.5, 11, 10.4, 3, 9.8]
Panther = [10.1, 11, 9.1, 20, 9.2, 10.8, 9.9, 9.2, 9.1, 9.1]

# CONVERT TO NUMPY ARRAYS  ----------------------------------------------------------
jaguar_array = np.array(Jaguar)
panther_array = np.array(Panther)

#  ------------------------- JAGUAR STATISTICS  -------------------------
print("Jaguar Statistics:")
print("Array:", jaguar_array)
print("Type:", type(jaguar_array))
print("Shape:", jaguar_array.shape)
print("Mean:", np.mean(jaguar_array))
print("Median:", np.median(jaguar_array))
print("Min:", np.min(jaguar_array))
print("Max:", np.max(jaguar_array))
print("25th Percentile:", np.percentile(jaguar_array, 25))
print("50th Percentile:", np.percentile(jaguar_array, 50))
print("75th Percentile:", np.percentile(jaguar_array, 75))
print("Variance:", np.var(jaguar_array))
print("Standard Deviation:", np.std(jaguar_array))
mode_result = stats.mode(jaguar_array, keepdims=True)
print("Mode:", mode_result.mode[0])

print("\n" + "-"*40 + "\n")

#  ------------------------- PANTHER STATISTICS  -------------------------
print("Panther Statistics:")
print("Array:", panther_array)
print("Type:", type(panther_array))
print("Shape:", panther_array.shape)
print("Mean:", np.mean(panther_array))
print("Median:", np.median(panther_array))
print("Min:", np.min(panther_array))
print("Max:", np.max(panther_array))
print("25th Percentile:", np.percentile(panther_array, 25))
print("50th Percentile:", np.percentile(panther_array, 50))
print("75th Percentile:", np.percentile(panther_array, 75))
print("Variance:", np.var(panther_array))
print("Standard Deviation:", np.std(panther_array))
mode_result = stats.mode(panther_array, keepdims=True)
print("Mode:", mode_result.mode[0])
```

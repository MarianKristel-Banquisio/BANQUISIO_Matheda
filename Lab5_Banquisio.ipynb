# Descriptive Statistics <hr style="border: none; border-top: 1px dashed #ff56bb;width: 75%; margin: 10px 0;"></hr>

```python
#jaguar vs panther
resistance = np.array([[10.6,9.1,9.3,9.8,10.5,10.4,9.5,11,10.4,3,9.8],
                       [10.1,11,9.1,20,9.2,10.8,9.9,9.2,9.1,9.1,np.nan]]).T
resistance
```

```python
#creating dataframe from numpy array
df = pd.DataFrame(resistance, columns=['Jaguar','Panther'])
df
```

```python
#summary of dataframe
df.info()
```

```python
type(df)
```

```python
#summary of statistics
df.describe()
```

```python
#median
jag_median = df['Jaguar'].median()
pan_median = df['Panther'].median()

print(f'Jaguar median: {jag_median: .2f}ohms')
print(f'Panther median: {pan_median: .2f}ohms')
```

```python
#mode
jag_mode = df['Jaguar'].mode()
pan_mode = df['Panther'].mode()
```

```python
jag_mode
```

```python
jag_mode.info()
```

```python
type(jag_mode)
```

```python
print('Jaguar mode:')
for i in jag_mode:
    print(i)
```

```python
print('Panther mode:')
for i in pan_mode:
    print(i)
```

```python
#Jaguar IQR 
Q1 = df['Jaguar'].quantile(0.25)
Q3 = df['Jaguar'].quantile(0.75)
print(f'Jaguar IQR = {Q3-Q1: .2f}ohms')
```

```python
#panther IQR
Q1 = df['Panther'].quantile(0.25)
Q3 = df['Panther'].quantile(0.75)
print(f'Panther IQR = {Q3-Q1: .2f}ohms')
```

# Bar Chart <hr style="border: none; border-top: 1px dashed #ff56bb;width: 75%; margin: 10px 0;"></hr>

```python
jag_mean = df['Jaguar'].mean()
pan_mean = df['Panther'].mean()

plt.figure(figsize=(7,4))
bars = plt.bar(['Jaguar','Panther'], [jag_mean, pan_mean], color=['blue','orange'])
plt.title('Average Resistance by machine')
plt.ylabel('Resistance (ohms)')

#bar labels
plt.bar_label(bars,padding=3)
plt.margins(y=0.1)
plt.show()
```

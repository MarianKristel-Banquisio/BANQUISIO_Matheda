```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r'C:\Users\Admin\Downloads\dataset\current-power.csv', delimiter=",")
df
```

```python
df.describe
```

```python
df.cov
```

```python
x = df['Current']
y = df['Power']

plt.Figure()
plt.scatter(x,y)
plt.title('Device Current vs Power Response')
plt.xlabel('Current (A)')
plt.ylabel('Power (W)')
plt.show()
```

```python
df.corr()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress (x,y)
line = slope * x + intercept

plt.figure(figsize= (7,7))

plt.scatter(x,y,
    color = "#F737A7",
    alpha = 0.7,
    label = 'Data Points')

plt.plot(x,line,
         color = "#22C8FF",
         label = f'Regression Line (r = {r_value:.3f})')

plt.title('Device Current vs Power Response')
plt.xlabel('Current (A)')
plt.ylabel('Power (W)')
plt.legend()
plt.show()
```

# NOx-Humidity <hr style="border: none; border-top: 1px dashed #ff56bb;width: 75%; margin: 10px 0;"></hr>

```python
air = pd.read_csv(r'C:\Users\Admin\Downloads\dataset\air-quality.csv', delimiter=",")
air
```

```python
# Regression Line
x = air['NOx']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope*x + intercept

#Scatter Plot
plt.figure(figsize= (7,4))

plt.scatter(x,y,
    color = "#F737A7",
    alpha = 0.2,
    label = 'Data Points')

# Line PLot
plt.plot(x,line,
         color = "#950404",
         label = f'Regression Line (r = {r_value:.3f})')

plt.title('NOx vs Humidity Correlation')
plt.xlabel('NOx (ug/cu.m)')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

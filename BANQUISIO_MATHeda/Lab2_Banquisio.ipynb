```python
ECE_Sit_Ins = ['Cyril', 'Glenn', 'Tracy', 'Marian', 'John','Paul', 'Renz', 'Kristel']

print("List of ECE Sit-Ins:")
for student in ECE_Sit_Ins:
    print(student)
```

```python
# with index
for i, student in enumerate(ECE_Sit_Ins, start=1):
    print(f"{i}. {student}")
```

```python
ECE_Sit_Ins = ['Cyril', 'Glenn', 'Tracy', 'Marian', 'John', 'Paul', 'Renz', 'Kristel']

# with section and grade
section = 'ECE-3AC'
grades = [88, 92, 95, 98, 85, 87, 93, 91]

# PRINT
print("-------------  ECE-3AC Student Records:  -----------\n")
for i in range(len(ECE_Sit_Ins)):
    print(f"Student: {ECE_Sit_Ins[i]}")
    print(f"Section: {section}")
    print(f"Grade: {grades[i]}")
    print("-" * 20)
```

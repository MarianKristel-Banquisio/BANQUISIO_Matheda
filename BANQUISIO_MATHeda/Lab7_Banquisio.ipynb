# Normal Distribution <hr style="border: none; border-top: 1px dashed #ff56bb;width: 75%; margin: 10px 0;"></hr>

```python
# Normal Distribution Curve
mean = df['Current'].mean()
std = df['Current'].std(ddof = 1)

# X-axis
x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
        color = "#9C1111",
        linewidth = 2)

plt.title('Normal Distribution Curve of Current Values',
          font = 'Palatino Linotype',
          fontsize = 12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
plt.show()
```

```python
plt.hist(df['Current'], 
         bins = 5,
         density = True,
         alpha = 0.8,
         color = "#ff56bb",
         edgecolor = "#000000")

plt.title('Histogram of Current Values',
          font = 'Times New Roman',
          fontsize = 12)
plt.xlabel('Current(mA)')
plt.ylabel('Frequency')

p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
        color = '#9C1111',
        linewidth = 2)

plt.show()
```

```python
plt.hist(df['Current'], 
         bins = 5,
         density = False,
         alpha = 0.8,
         color = "#ff56bb",
         edgecolor = "#000000")

plt.title('Histogram of Current Values',
          font = 'Times New Roman',
          fontsize = 12)
plt.xlabel('Current(mA)')
plt.ylabel('Frequency')

p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
        color = '#012413',
        linewidth = 2)

plt.show()
```

```python
# X-axis
x_min = df['Z-Score'].min()
x_max = df['Z-Score'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, df['Z-Score'].mean(),
                   df['Z-Score'].std(ddof = 1))

plt.plot(x, p,
        color = '#126782',
        linewidth = 2)

plt.title('Z-Score',
          font = 'Times New Roman',
          fontsize = 12)
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')

plt.plot(x, p,
        color = '#E09406',)

plt.show()
```

# Measures of Central Tendency <hr style="border: none; border-top: 1px dashed #ff56bb;width: 75%; margin: 10px 0;"></hr>

```python

 # Jaguar vs. Panther
resistance = np.array([[10.6,9.1,9.3,9.8,10.5,10.4,9.5,11,10.4,3,9.8],
                       [10.1,11,9.1,20,9.2,10.8,9.9,9.2,9.1,9.1,np.nan]])
resistance
```

```python
np.shape(resistance)
```

```python
resistance.T
```

```python
np.mean(resistance,axis=0)
```

```python
np.mean(resistance,axis=1)
```

```python
# Mean
np.nanmean(resistance,axis=1)
```

```python
jaguar_mean = np.nanmean(resistance, axis=1)[0]
panther_mean = np.nanmean(resistance, axis=1)[1]

print(f'Jaguar Mean: {jaguar_mean: .2f} ohms')
print(f'Panther Mean: {panther_mean: .2f} ohms')
```

```python
# Median
np.nanmedian(resistance, axis=1)
```

```python
jag_median = np.nanmedian(resistance, axis=1)[0]
pan_median = np.nanmedian(resistance, axis=1)[1]

print(f'Jaguar Median: {jag_median: .2f} ohms')
print(f'Panther Median: {pan_median: .2f} ohms')
```

```python
# Mode
stats.mode(resistance, axis=1)
```

```python
my_array = [1,1,2,2,3,4,5]
multimode(my_array)
```

```python
jag_res = resistance[0]

print("jaguar mode")
for i in multimode(jag_res):
    print(f'{i} ohms')
```

```python
pan_res = resistance[1]

print("panther mode")
for i in multimode(pan_res):
    print(f'{i} ohms')
    
```

# Measures of Variability <hr style="border: none; border-top: 1px dashed #ff56bb;width: 75%; margin: 10px 0;"></hr>

```python
# Jaguar vs. Panther
resistance = np.array([[10.6,9.1,9.3,9.8,10.5,10.4,9.5,11,10.4,3,9.8],
                       [10.1,11,9.1,20,9.2,10.8,9.9,9.2,9.1,9.1,np.nan]])
resistance
```

```python
# Range
res_max = np.nanmax(resistance, axis=1)
res_min = np.nanmin(resistance, axis=1)

res_range = res_max - res_min
res_range
```

```python
print(f'Jaguar Range: {res_range[0]: .2f} ohms')
print(f'Panther Range: {res_range[1]: .2f} ohms')
```

```python
# 1st Quartile
res_Q1 = np.nanpercentile(resistance, 25, axis=1)
res_Q1
```

```python
# 2nd Quartile
res_Q2 = np.nanpercentile(resistance, 50, axis=1)
res_Q2
```

```python
# 3rd Quartile
res_Q3 = np.nanpercentile(resistance, 75, axis=1)
res_Q3
```

```python
# IQR
print(f'Jaguar IQR: {res_Q3[0] - res_Q1[0]: .2f} ohms')
print(f'Panther IQR: {res_Q3[1] - res_Q1[1]: .2f} ohms')
```

```python
# Variance
res_variance = np.nanvar(resistance,ddof=1, axis=1)
res_variance
```

```python
print(f'Jaguar Variance: {res_variance[0]: .2f} ohms')
print(f'Panther Variance: {res_variance[1]: .2f} ohms')
```

```python
# Standard Deviation
res_std = np.nanstd(resistance,ddof=1, axis=1)
res_std
```

```python
print(f'Jaguar Standard Deviation: {res_std[0]: .2f} ohms')
print(f'Panther Standard Deviation: {res_std[1]: .2f} ohms')
```
